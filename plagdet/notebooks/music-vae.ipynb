```python
import tensorflow as tf
import note_seq
from magenta.models.music_vae import configs
from magenta.models.music_vae.trained_model import TrainedModel

# Load the MusicVAE model
config = configs.CONFIG_MAP[MUSICVAE_2BAR_BIG_CONFIG]
model = TrainedModel(config, batch_size=4, checkpoint_dir_or_path=MUSICVAE_2BAR_BIG_PATH)

def midi_to_musicvae_vectors(midi_file_path):
    # Load and preprocess the MIDI file
    try:
        ns = note_seq.midi_file_to_note_sequence(midi_file_path)
    except Exception as e:
        print(f"Error loading MIDI file: {e}")
        return None

    # Convert NoteSequence to MelodyTensors
    try:
        melodies = note_seq.sequences_lib.extract_subsequences(ns, config.hparams.max_seq_len)
        melody_tensors = config.data_converter.to_tensors(melodies).inputs
    except Exception as e:
        print(f"Error converting NoteSequence to MelodyTensors: {e}")
        return None

    # Encode the melodies
    try:
        z, _, _ = model.encode(melody_tensors)
        return z
    except Exception as e:
        print(f"Error encoding melodies: {e}")
        return None

# Example usage
midi_file = os.path.join(LAKH_CLEAN_PATH, 'Led_Zeppelin', 'Stairway_To_Heaven.2.mid')
encoded_vectors = midi_to_musicvae_vectors(midi_file)

if encoded_vectors is not None:
    print(f"Encoded vectors shape: {encoded_vectors.shape}")
    print(f"First encoded vector: {encoded_vectors[0]}")
else:
    print("Failed to encode MIDI file.")
```
