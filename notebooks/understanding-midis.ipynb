## does the length of a bar in seconmds change when we use the magenta extraction?


```python
files = [
    r'C:\Users\cunn2\OneDrive\DSML\Project\thesis-repo\plagdet\data\midi_databases\lakh_clean\Wonder_Stevie\Isnt_She_Lovely.1.mid',
    r'C:\Users\cunn2\OneDrive\DSML\Project\thesis-repo\plagdet\data\midi_databases\aa_midis\No-Scrubs.mid'
]

extractor = MonophonicMelodyExtractor()

for file in files:
    print(f"Processing file: {file}")
    
    # Calculate the length of a bar before extraction
    bpm, time_signature = get_tempo_and_time_signature(file)
    beats_per_bar = time_signature[0]
    seconds_per_beat = 60 / bpm
    bar_length_before = beats_per_bar * seconds_per_beat
    print(f"Length of a bar before extraction: {bar_length_before:.2f} seconds")
    
    # Apply the make_file_valid method
    extractor.make_file_valid(file, 'temp_extracted.mid')
    
    # Calculate the length of a bar after extraction
    bpm_after, time_signature_after = get_tempo_and_time_signature('temp_extracted.mid')
    beats_per_bar_after = time_signature_after[0]
    seconds_per_beat_after = 60 / bpm_after
    bar_length_after = beats_per_bar_after * seconds_per_beat_after
    print(f"Length of a bar after extraction: {bar_length_after:.2f} seconds")
    
    # Calculate and print the difference
    difference = abs(bar_length_after - bar_length_before)
    print(f"Difference in bar length: {difference:.2f} seconds")
    
    print("\n")  # Add a newline for readability between files

# Clean up the temporary file
if os.path.exists('temp_extracted.mid'):
    os.remove('temp_extracted.mid')

```

```python
def count_bars(midi_file):
    mid = mido.MidiFile(midi_file)
    ticks_per_beat = mid.ticks_per_beat
    time_sig = None
    total_ticks = 0

    for msg in mid:
        if msg.type == 'time_signature':
            time_sig = msg
        total_ticks += msg.time

    if not time_sig:
        raise ValueError("No time signature found")

    beats_per_bar = time_sig.numerator
    total_beats = total_ticks / ticks_per_beat
    total_bars = total_beats / beats_per_bar

    return total_bars
```
